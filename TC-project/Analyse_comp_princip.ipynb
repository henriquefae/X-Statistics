## Partie I: Etude théorique de l'analyse en composantes principales

Soit $(\Omega , \mathcal A, \mathbb{P})$ un espace de probabilités. Soit $D\in \mathbb{N}^*$ et $X$ un vecteur aléatoire à valeurs dans $\mathbb{R}^D$ dont toutes les composantes sont des variables aléatoires de carré intégrable. Il faut interpréter $D$, la dimension de $X$, comme le nombre de caractéristiques d'un système d'intérêt, qui sera très grand en pratique dans les cas qui nous intéresseront par la suite. 

L'analyse en composantes principales a pour but de rechercher une approximation du vecteur aléatoire $X$ qui sera construite à l'aide d'un vecteur aléatoire $Y$ à valeurs dans $\mathbb{R}^d$ dont la dimension $d\in \mathbb{N}^*$ a vocation à être beaucoup plus faible que $D$ la dimension initiale du vecteur $X$. On notera $X^i$ la $i^{eme}$ composante du vecteur $X$ pour $1\leq i \leq D$ et $Y^j$ la $j^{eme}$ composante du vecteur $Y$ pour $1\leq j \leq d$. 

Pour ce faire, l'analyse en composantes principales consiste à rechercher une famille orthonormale **déterministe** $(x_1, \ldots, x_d)$ de $\mathbb{R}^D$ (muni du produit scalaire euclidien usuel qu'on notera $\langle \cdot, \cdot \rangle$) de telle sorte que la quantité

\begin{equation}
F_X(x_1, \ldots, x_d) = \inf_{(Y^1, \ldots, Y^d) \in (\mathbb{L}^2_{\mathbb{P}})^d} \mathbb{E}\left[ \left\|X - \sum_{j=1}^d Y^j x_j\right\|^2\right],
\end{equation}

où $\|\cdot\|$ désigne la norme euclidienne de $\mathbb{R}^D$, soit la plus petite possible.

T1) Soit $(x_1,\ldots,x_d)$ une famille orthonormale déterministe de $\mathbb{R}^D$. Montrer que le problème de minimisation précédent admet un unique minimiseur et en donner une expression en fonction de $X$ et $(x_1, \ldots, x_d)$.

Réponse T1)
$$
F_X(x_1, \ldots, x_d) = \inf_{\vec{Y} \in (L^2)^d} \mathbb{E} \left[ \left\| X - \sum_{j=1}^{d} Y^{j} x_j \right\|^2 \right]
$$

On peut compléter la famille $ (x_1, \ldots, x_d) $ en une base orthonormée $ (x_1, \ldots, x_D) $ de $ \mathbb{R}^D $.

Ainsi :

$$
X = \sum_{j=1}^{D} \langle X, x_j \rangle x_j
$$

Donc :

$$
X - \sum_{j=1}^{d} Y^{j} x_j = \sum_{j=1}^{d} (\langle X, x_j \rangle - Y^{j})x_j + \sum_{j=d+1}^{D} \langle X, x_j \rangle x_j
$$

Par orthonormalité, on a :

$$
\| X - \hat{X} \|^2 = \sum_{j=1}^{d} |\langle X, x_j \rangle - Y^{j}|^2 + \sum_{j=d+1}^{D} | \langle X, x_j \rangle |^2
$$
Avec 
$$
\hat{X} \ = \sum_{j=1}^{d} Y^{j} x_j
$$
Par linéarité de E, on veut minimiser 
$$
\sum_{j=1}^{d} \mathbb{E}\left[|\langle X, x_j \rangle - Y^{j}|^2\right]
$$

D'où on obtient :

$$
Y^{j} = \langle X, x_j \rangle
$$

On trouve donc que la meilleure approximation de X par ${Vect}(x_1, \ldots, x_d)$ est sa projection orthogonale, qui est uniquement déterminée par :

$$
\hat{X} = \sum_{j=1}^{d} \langle X, x_j \rangle x_j

T2) Montrer que 

\begin{equation}
F_X(x_1, \ldots, x_d) = -\sum_{j=1}^d x_j^T A_X x_j + c_X
\end{equation}

avec $A_X\in \mathbb{R}^{d\times d}$ la matrice symétrique définie positive et $c_X\in \mathbb{R}$ donnés par 

\begin{equation}
A_X = \mathbb{E}\left[ XX^T\right] \quad \text{et} \quad c_X = \mathbb{E}[\|X\|^2].
\end{equation}

Montrer en particulier que $c_X = \mathrm{Tr}\,A_X$.

Réponse T2)


$$
\|X - \hat{X} \|^2 = \|X\|^2 - \sum_{j=1}^d \langle X, x_j \rangle^2 = \|X\|^2 - \sum_{j=1}^d (x_j^T X)^2
$$

Donc :

$$
\mathbb{E} \left[ \| X - \hat{X} \|^2 \right] = \mathbb{E} \left[ \|X\|^2 \right] - \sum_{j=1}^d \mathbb{E} \left[ (x_j^T X)^2\right]
$$

Or :

$$
\mathbb{E} \left[ (x_j^T X)^2\right] = x_j^T \, \mathbb{E}(X X^T) \, x_j = x_j^T A_X x_j
$$

Donc :

$$
F_X(x_1, \ldots, x_d) = c_X - \sum_{j=1}^d x_j^T A_X x_j
$$

où :

$$
c_X = \mathbb{E}(\|X\|^2), \quad A_X = \mathbb{E}(X X^T)
$$

Et on note aussi :

$$
c_X = \mathbb{E}(\|X\|^2) = \mathbb{E} \left[ \text{tr}(X X^T) \right] = \text{tr}(A_X)
$$


Nous allons à présent supposer que $d=1$ pour les questions T3) à T6)

T3) On considère alors le problème de minimisation
$$
\mathop{\inf}_{x_1\in \mathbb{R}^D, \; \|x_1\| = 1} F_X(x_1). 
$$
Montrer que l'ensemble des minimiseurs de ce problème de minimisation est l'ensemble des vecteurs propres de norme $1$ associés à la plus grande valeur propre de la matrice $A_X$. En déduire la valeur de l'infimum considéré en fonction de $\lambda_1$ la plus grande valeur propre de $A_X$. 


Réponse T3)

Pour $d = 1$, on a :

$$
F_X(x_1) = c_X - x_1^T A_X x_1
$$

où :

$$
A_X = \mathbb{E}(X X^T), \quad (A_X)_{ij} = \mathbb{E}(X^i X^j) = (A_X)_{ji}
$$

et :

$$
(A_X)_{i,i} = \mathbb{E}((X^i)^2) \geq 0
$$

Alors $A_X$ est symétrique réelle, donc elle est diagonalisable par une base orthonormée, selon le théorème spectral.

Pour $v \in \mathbb{R}^D$, on a :

$$
v^T A_X v = \mathbb{E} \left[ (v^T A_X)^2 \right] \geq 0
$$

On peut écrire alors :

$$
A_X = U^T D U, \quad \text{où} \quad (D)_{i,j} =
\begin{cases}
\lambda_i & \text{si } i = j \\
0 & \text{sinon}
\end{cases}
\quad \text{avec } \lambda_i \geq 0 \text{ (valeurs propres)}
$$

et $U^{-1} = U$ 

Alors :

$$
v^T A_X v = (U v)^T D (U v) = w^T D w = \sum_{i=1}^D \lambda_i (w^i)^2
$$

et :

$$
\|w\| = ( (U v)^T (U v) )^{1/2} = \|v\|
$$

Donc :

$$
F_X(v) = c_X - \sum_{j=1}^{D} \lambda_j (w^j)^2, \quad \text{où } w^j = (U v)^j
$$

et ainsi :

$$
F_X(v) \geq c_X - \left( \max_{j = 1, \ldots, D} \lambda_j \right) \|v\|^2
$$

On considère $\lambda_1 = \left( \max_{j = 1, \ldots, D} \lambda_j \right) $. Alors, pour tout $ v \in \ker(A_X - \lambda_1 I_D)$, on a :

$$
v^T A_X v = \lambda_1 \|v\|^2 \quad \Rightarrow \quad F_X(v) = c_X - \lambda_1 \|v\|^2
$$

Donc :

$$
\inf_{\|x_1\| = 1} F_X(x_1) = c_X - \lambda_1 \quad \text{où } x_1 \in \ker(A_X - \lambda_1 I_D)
$$


A présent, soit $(X_n)_{n\in \mathbb{N}^*}$ une suite de variables aléatoires iid de même loi que $X$. Pour tout $N\in \mathbb{N}^*$, on considère la matrice aléatoire 
$\displaystyle A^N:= \frac{1}{N}\sum_{i=1}^N X_i X_i^T$. Il est facile de voir que $A^N$ est presque sûrement une matrice symétrique et on note $\Lambda_1^N$ sa plus grande valeur propre (qui est alors une variable aléatoire à valeurs dans $\mathbb{R}$). 

T4) Montrer que $(A^N)_{N\in \mathbb{N}^*}$ converge presque sûrement vers $A_X$ lorsque $N$ tend vers $+\infty$. 

Réponse T4)


Soit $(X_n)_{n \in \mathbb{N}} $ une suite de vecteurs aléatoires i.i.d.,   $ X_i \in \mathbb{R}^D $

Pour chaque composante $(k,l)$ de $ X_i X_i^T, i \in \mathbb{N}$, on a :

$$
\mathbb{E} \left[ (X_i X_i^T)_{k,\ell} \right] = (A_X)_{k,\ell}
$$

Ainsi, d’après la loi forte des grands nombres, on a :

$$
\left( \frac{1}{N} \sum_{i=1}^{N} X_i X_i^T \right)_{k,\ell} \xrightarrow{\text{p.s.}} (A_X)_{k,\ell}, \quad \forall (k,\ell) \in \{1, \ldots, D\}^2
$$

Alors :

$$
A^N \xrightarrow{\text{p.s.}} A_X \quad \text{lorsque } N \to \infty
$$

T5) Montrer que presque sûrement $\displaystyle \lambda_1 \leq \mathop{\liminf}_{N\to +\infty} \Lambda_1^N$. 

Réponse T5) 

La plus grande valeur propre de $A^N$ peut s’exprimer comme :

$$
\Lambda_1^N = \sup_{\|v\| = 1} v^T A^N v
$$

Pour tout vecteur  $v$ unitaire, on a :

$$
v^T A^N v \xrightarrow{\text{p.s.}} v^T A_X v
$$

Si $v \in \ker(A_X - \lambda_1 I_D)$, alors :

$$
v^T A^N v \xrightarrow{\text{p.s.}} v^T A_X v = \lambda_1
$$

Comme $\Lambda_1^N \geq v^T A^N v$ pour tout $N \in \mathbb{N}^*$, alors :

$$
\liminf_{N \to \infty} \Lambda_1^N \geq \lambda_1 \quad \text{p.s.}
$$

T6) Montrer que $\displaystyle \lambda_1 \geq \mathop{\limsup}_{N\to +\infty} \Lambda_1^N$ presque sûrement. En déduire que $(\Lambda_1^N)_{N\in \mathbb{N}^*}$ converge presque sûrement vers $\lambda_1$ lorsque $N$ tend vers $+\infty$. 

Réponse T6) 
On peut définir :

$$
\Lambda_1^K = \sup_{\|v\| = 1} v^T A^N v
$$

Soit $\Lambda_1^K = \sup_{N \in \mathbb{N}} \Lambda_1^N$, $K \in \mathbb{N}^*$, et $v \in \ker(A^K - \Lambda_1^K I_D)$

Si $\|v\| = 1$, alors :

$$
\Lambda_1^K \geq \Lambda_1^N \geq v^T A^N v \xrightarrow{\text{p.s.}} v^T A_X v \leq \lambda_1
$$

Donc :

$$
\limsup_{N \to \infty} \Lambda_1^N \leq \lambda_1 \quad \text{p.s.}
$$

Comme $\limsup_{N \to \infty} \Lambda_1^N \leq \lambda_1 \leq \liminf_{N \to \infty} \Lambda_1^N $ et $\limsup_{N \to \infty} \Lambda_1^N \geq \liminf_{N \to \infty} \Lambda_1^N $ presque sûrement, alors :

$$
\left( \Lambda_1^N \right)_{N \in \mathbb{N}} \xrightarrow{\text{p.s.}} \lambda_1

On admettra dans la suite de l'énoncé que ce résultat se généralise au cas où $d$ est quelconque. Plus précisément, si on note $\lambda_1 \geq \lambda_2 \geq \ldots \geq \lambda_D$ les $D$ valeurs propres de la matrice $A_X$ (en comptant leur multiplicité) et si $\lambda_d < \lambda_{d+1}$, l'ensemble des minimiseurs de $F(x_1,\ldots,x_d)$ est exactement l'ensemble des familles orthonormales de vecteurs $(x_1, \ldots, x_d)$ de $\mathbb{R}^D$ telles que 

\begin{equation}
\operatorname{Vect}\{x_1, \ldots, x_d\} = \operatorname{Ker}(A_X - \lambda_1 \mathrm{Id}) \oplus \cdots \oplus \operatorname{Ker}(A_X - \lambda_d \mathrm{Id})
\end{equation}

En pratique, cependant, on ne connaît pas exactement la valeur de la matrice $A_X$. Il est beaucoup plus courant que l'on ait accès à un nombre $N\in \mathbb{N}^*$ de réalisations $X_1, \ldots, X_N$ de vecteurs aléatoires indépendants de même loi que $X$. On calcule alors plutôt $U_1^N, \ldots, U_d^N$ une famille orthonormée de vecteurs propres de la matrice $A^N$ de telle sorte que

\begin{equation}
A^N U_j^N = \Lambda_j^N U_j^N \quad \forall 1\leq j \leq d
\end{equation}

avec $\Lambda_1^N \geq \ldots \geq \Lambda_d^N$ les $d$ plus grandes valeurs propres de la matrice $A^N$.

T7) Donner l'expression de l'unique minimiseur du problème de minimisation suivant en fonction de $(X_i)_{1\leq i \leq N}$ et $(U_j^N)_{1\leq j \leq d}$:

$$
 \mathop{\rm inf}_{(Y_i^1, \ldots, Y_i^d)_{1\leq i \leq N} \in ((\mathbb{L}^2_{\mathbb{P}})^d)^N}  \sum_{i=1}^N\left[ \left\|X_i - \sum_{j=1}^d Y_i^j U_j^N\right\|^2\right]
$$

Réponse T7)

Soit :

$$
f(Y_1, \ldots, Y_N) = \sum_{i=1}^{N} \|X_i- \sum_{j=1}^{d} Y_i^jU_j^N\|^2 
$$

avec $Y_i \in \mathbb{R}^d$,  $U_j^N \in \mathbb{R}^D$

Pour chaque $X_i \in \mathbb{R}^D$, on veut minimiser sa distance à l’espace :

$$
\text{Vect}(U_1^N, \ldots, U_d^N) = \langle U^N \rangle
$$

On procède par compléter la famille $\{ U_1^N, \ldots, U_d^N \}$ pour créer une base orthonormée de  $\mathbb{R}^D$ :

$$
\{ U_1^N, \ldots, U_D^N \}
$$

Ainsi :

$$
X_i = \sum_{j=1}^{D} \langle X_i, U_j^N \rangle U_j^N
$$

Donc :

$$
\|X_i - \sum_{j=1}^{d} Y_i^j U_j^N \|^2 
$$

$$
= \left\| \sum_{j=1}^{d} (\langle X_i, U_j^N \rangle - Y_i^j) U_j^N + \sum_{j=d+1}^{D} \langle X_i, U_j^N \rangle U_j^N \right\|^2
$$



Par orthonormalité des vecteurs $U_j^N$, cela donne :

$$
 \sum_{j=1}^{d} (\langle X_i, U_j^N \rangle - Y_i^j)^2 + \sum_{j=d+1}^{D} \langle X_i, U_j^N \rangle^2
$$

Donc :

$$
f(Y_1, \ldots, Y_N) \geq \sum_{i=1}^{N} \sum_{j=d+1}^{D} \langle X_i, U_j^N \rangle^2
$$

D’où, on obtient l’égalité uniquement pour :

$$
Y_i^j = \langle X_i, U_j^N \rangle, \quad \forall i \in \{1, \ldots, N\}, \quad \forall j \in \{1, \ldots, d\}
$$

De plus, soit :

$$
\hat{X}_i^N = \sum_{j=1}^{d} \langle X_i, U_j^N \rangle U_j^N
$$

la projection de $X_i$. 

Alors :

$$
\|X_i - \hat{X}_i^N\|^2 = \|X_i\|^2 - \sum_{j=1}^{d} \langle X_i, U_j^N \rangle^2 = \|X_i\|^2 - \sum_{j=1}^{d} (U_j^N)^T X_i X_i^T U_j^N
$$

Donc :

$$
\sum_{i=1}^{N} \|X_i - \hat{X}_i^N\|^2 = \sum_{i=1}^{N} \|X_i\|^2 - N\sum_{j=1}^{d} (U_j^N)^T \left( \frac{1}{N} \sum_{i=1}^{N} X_i X_i^T \right) U_j^N
$$

Soit :


On obtient donc :

$$
\sum_{i=1}^{N} \|X_i - \hat{X}_i^N\|^2 = \sum_{i=1}^{N} \|X_i\|^2 - N \sum_{j=1}^{d} \Lambda_j^N
$$

Ainsi, l’infimum est donné par :

$$
\sum_{i=1}^{N} \|X_i\|^2 - N \sum_{j=1}^{d} \Lambda_j^N
$$

Pour tout $1\leq i \leq N$, il est alors tentant que considérer que $X_i$ peut être approché par la quantité $\sum_{j=1}^d Y_i^j U_j^N$. L'analyse en composantes principales consiste alors à plutôt analyser les quantités $(Y_i^j)_{1\leq i \leq N, 1\leq j \leq d}$ qui permet d'obtenir une représentation de l'information contenue dans la valeur des réalisations $(X_i)_{1\leq i \leq N}$ de plus petite dimenson ($dN$ au lieu de $DN$) et souvent plus aisée à interpréter.

On a utilisé la base de données "Wisconsin Breast Cancer Dataset", comprenant 569 cas de tumeurs du sein répartis en 357 tumeurs bénignes et 212 malignes. Chaque tumeur est caractérisée par 30 variables numériques décrivant des attributs tels que la taille, la forme et la texture.

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
XX = StandardScaler().fit_transform(x) # normalizing the features
XX.view()
```

P2) Commenter ce que fait la fonction StandardScaler().fit_transform(x) ci-dessus et ce que représente le tableau XX.

Réponse P2)

La fonction StandardScaler().fit_transform(x) réalise une standardisation des données qui fait chaque colonne de XX avoir une moyenne de 0 et un écart-type de 1.

\begin{align*}
\text{fit-transform(x)} &: \quad 
x_{ij}^{\text{scaled}} = \frac{x_{ij} - \mu_j}{\sigma_j},\quad où \quad \mu_j = \frac{1}{n} \sum_{i=1}^n x_{ij}, \quad 
\sigma_j = \sqrt{\frac{1}{n} \sum_{i=1}^n (x_{ij} - \mu_j)^2}
\end{align*}

$x_{ij}$ est la valeur de la j-ème caractéristique pour la i-ème observation


Le tableau XX a les mêmes dimensions que x (569×30), mais avec des valeurs standardisées. Ainsi, chaque colonne correspond à une caractéristique originale (ex: mean radius), mais centrée-réduite. 


Ce processus permettra l'ACP d'utiliser la matrice de covariance, qui dépend des unités de mesure. Sans standardisation, une caractéristique avec des valeurs larges (ex: area) dominerait artificiellement l'analyse.

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
```

Nous allons à présent effectuer une analyse en composantes principales à partir des données renormalisées contenues dans le tableau de données XX. Ici, nous sommes dans un cas où $D = 30$ et $N = 569$.

P3) Construire la matrice $A^N\in \mathbb{R}^{D\times D}$ définie juste avant la question T4).  

```python
#Definition de la matrice A_N
A_N = (1 / XX.shape[0]) * XX.T @ XX

print("Dimensions de A_N :", A_N.shape)
print("Aperçu des 5 premières lignes/colonnes :\n", A_N[:5, :5])
```

Définition: La matrice $A^N$ est la matrice de covariance empirique des données standardisées $XX\in \mathbb{R}^{N\times D}$ (où N = 569, D = 30)

Importance: $A^N$ capture les relations linéaires entre les 30 caractéristiques standardisées et ses valeurs propres/vecteurs propres serviront à déterminer les composantes principales.

P4) Tracer les valeurs propres de la matrice $A$ par ordre décroissant en échelle log. Commenter.

```python
# Calcul des valeurs propres et vecteurs propres
eigenvalues, eigenvectors = np.linalg.eig(A_N)

# Tri des valeurs propres et vecteurs propres par ordre décroissant
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[sorted_indices]
eigenvectors_sorted = eigenvectors[:, sorted_indices]

# Tracé en échelle logarithmique
plt.figure(figsize=(10, 6))
plt.plot(range(1, 31), np.log10(eigenvalues_sorted), 'bo-', markersize=5)
plt.xlabel('Rang de la valeur propre', fontsize=12)
plt.ylabel(r'log10($\lambda_i$)', fontsize=12)
plt.title('Valeurs propres de $A^N$ (échelle log)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(range(1, 31))
plt.show()
```

Commentaire: 

Les valeurs propres, triées par ordre décroissant, révèlent une décroissance rapide : les premières valeurs propres dominent largement, indiquant que les premières composantes principales capturent l'essentiel de la variance. Cela suggère la présence de redondances ou de bruit dans les caractéristiques originales. Alors, cette propriété permet une réduction de dimension efficace par l'ACP, où seules quelques composantes suffisent à conserver l'information pertinente.

```python
# Extraction des deux premiers vecteurs propres
U1_N = eigenvectors_sorted[:, 0] 
U2_N = eigenvectors_sorted[:, 1] 

# Affichage des résultats
print("Vecteur propre U1_N (associé à lambda_1 = {:.3f}):\n{}".format(eigenvalues_sorted[0], U1_N))
print("\nVecteur propre U2_N (associé à lambda_2 = {:.3f}):\n{}".format(eigenvalues_sorted[1], U2_N))
```
